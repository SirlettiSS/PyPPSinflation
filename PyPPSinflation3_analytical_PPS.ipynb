-----

The $\varepsilon_i$ and $H$ are found to be analytically

$H^2 \simeq \frac{1}{3M^2_{Pl}} V$

$\varepsilon_1 \simeq \frac{M^2_{Pl}}{2} \left(\frac{V'}{V}\right)^2$

$\varepsilon_2 \simeq 2M^2_{Pl}\left[\left(\frac{V'}{V}\right)^2-\frac{V''}{V}\right]$

$\varepsilon_2 \varepsilon_3 \simeq 2M^4_{Pl}\left[\frac{V''' V'}{V^2}-\frac{3V'' V'^2}{V^3} +2\left(\frac{V'}{V}\right)^4\right]$

 
-----

For the background and its relation with the e-folds number, we have: 

$N_{end}- N \simeq -\frac{1}{M^2_{Pl}} \int_{\phi}^{\phi_{end}} \frac{V}{V'} d\phi$,

where we define $\phi_{end}$ to be the value when $\varepsilon_1=1$. 

-----

We want to follow these steps:

1. Define the potential, i.e., define the model.

2. Compute $V$, $V'$, $V''$, $\varepsilon_1$, $\varepsilon_2$, $\varepsilon_3$, and $\varepsilon_4$.

3. Compute $\phi_{end}$ from $\varepsilon_1(\phi_{end})=1$.

4. Compute $\Delta N$.

5. Invert the result from the previous step to find $\phi(\Delta N)$.

6. Rewrite the expressions in terms of $\Delta N$.

7. Compute $a_{S0,1,2}$ and $a_{T0,1,2}$.

8. Compute $\mathcal{P}(k)$.

Note that at first order, terms like $\varepsilon^2_1, \varepsilon^2_2, \varepsilon_2\varepsilon_3$, etc., are set to zero, simplifying the equations for $a_{S/T}$. At second order, the equations are complete. In Vennin's paper, the approach differs as he modifies the equations for $H$ and $\varepsilon_i$.

Steps 1 to 6 should be computed analytically by hand, or at most by using analytical software such as Mathematica or Maple. The PyPPSinflation package includes a Wolfram Mathematica notebook, Quantities_wrt_N, which can compute steps 1 to 6 analytically for any inflation model. Steps 7 and 8 are computed in this Jupyter notebook.

## An example of analytical computation with the simplest model: $\frac{1}{2}m^2 \phi^2$.

In this notebook, we use the KKLT model as an example. Since its equations are too lengthy, we also present the simplest model of inflation for computation examples.

For the $V(\phi) = \frac{1}{2}m^2\phi^2$ model, performing all the calculations (which are simple enough to be done by hand) yields:


$H^2 = \frac{m^2 \phi^2}{6 M^2_{Pl}}$

$\varepsilon_1 =\frac{ 2M^2_{Pl}}{\phi^2}$

$\varepsilon_2 =\varepsilon_3= 2\varepsilon_1$

From the equation $\varepsilon_1(\phi_{end}) =1$ we get $\phi^2_{end} = 2 M^2_{Pl}$, and integrating the $\Delta N$ and inverting it we get

$\phi^2 = 4 M^2_{Pl} \left(\Delta N + \frac{1}{2} \right).$

Substituting we have

$\varepsilon_1  = \frac{\varepsilon_2}{2}= \frac{\varepsilon_3}{2}=\frac{1}{2\Delta N +1} \ \ \ and \ \ \ H^2 = \frac{2}{3} m^2(\Delta N+\frac{1}{2})$,

and then we are able to easely compute the PPS. 

## The choice of the coupling constant

In the first notebook of PyPPSinflation, which is dedicated to numerical computation, we explain the procedure for normalization, specifically the choice of the coupling constant. Here, we focus on the second step of that procedure, which concerns normalization for the analytical part.

Let's recall some important points:

We choose the coupling constant of the theory, which in this case is the mass, by imposing that the scalar primordial power spectrum at the pivot scale has the value $\mathscr{P}_{R}(k_*) = a_0 \mathscr{P}_{0R}(k_*) = 2\times 10^{-9}$.

To achieve the same value of the PPS at the pivot scale for both numerical and analytical computations, it is possible to use different coupling constants.

At the end of this notebook, we will address the normalization of the analytical coupling constant(s).

```python
#Defining the first order coefficients

C = -0.7296

def as01(N): 
    return 1-2*(C+1)*eps1(N) - C*eps2(N)

def as11(N): 
    return -2*eps1(N) - eps2(N) 

def as21(N):
    return 0

def at01(N):
    return 1-2*(C+1)*eps1(N) 

def at11(N):
    return -2*eps1(N)

def at21(N):
    return  0
```

```python
#Defining the a parameters at the second order 

C = -0.7296

def as02(N): 
    return 1-2*(C+1)*eps1(N) - C*eps2(N) + (2*C**2 +2*C + 0.5*np.pi**2 -3)*eps1(N)**2 + (C**2 - C + 7/12* np.pi**2 -6)*eps1(N)*eps2(N) + (0.5*C**2 + np.pi**2/8 -1)*eps2(N)**2 + (-0.5*C**2 + np.pi**2/24)*eps2(N)*eps3(N)

def as12(N): 
    return -2*eps1(N) - eps2(N) + 2*(2*C+1)*eps1(N)**2 + (2*C-1)*eps1(N)*eps2(N) + C*eps2(N)**2 - C*eps2(N)*eps3(N)

def as22(N):
    return 4*eps1(N)**2 +2*eps1(N)*eps2(N)+eps2(N)**2-eps2(N)*eps3(N)

def at02(N):
    return 1-2*(C+1)*eps1(N) + (2*C**2 +2*C + 0.5*np.pi**2 -3)*eps1(N)**2 + (-C**2 -2*C + np.pi**2/12 -2)*eps1(N)*eps2(N)

def at12(N):
    return -2*eps1(N) +2*(2*C+1)*eps1(N)**2- 2*(C+1)*eps1(N)*eps2(N)

def at22(N):
    return  4*eps1(N)**2 - 2*eps1(N)*eps2(N)
```

```python
#Defining the a parameters at the third order 
#The scalar coefficients are defined so that it is possible to set eps4(N) = 0

C = -0.7296

def as03(N, e4 = True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4 = 0
    return (1 - 2 * (1 + C) * eps1(N) + (-3 + 2 * C + 2 * C**2 + (np.pi**2) / 2) * eps1(N)**2 - C * eps2(N) + (-6 - C + C**2 + (7 * np.pi**2) / 12) * eps1(N) * eps2(N) + (1 / 8) * (-8 + 4 * C**2 + np.pi**2) * eps2(N)**2 + (1 / 24) * (-12 * C**2 + np.pi**2) * eps2(N) * eps3(N) - (1 / 24) * (-16 - 24 * C + 4 * C**3 + 3 * C * np.pi**2 + 14 * zeta(3)) * (8 * eps1(N)**3 + eps2(N)**3) + (1 / 12) * (72 * C + 36 * C**2 + 13 * np.pi**2 - 8 * C * np.pi**2 - 84 * zeta(3)) * eps1(N)**2 * eps2(N) - (1 / 24) * (16 - 24 * C - 12 * C**2 + 8 * C**3 - 15 * np.pi**2 + 6 * C * np.pi**2 + 100 * zeta(3)) * eps1(N) * eps2(N)**2 + (1 / 24) * (16 - 4 * C**3 + C * np.pi**2 - 8 * zeta(3)) * (eps2(N) * eps3(N)**2 + eps2(N) * eps3(N) * epsi4) + (1 / 24) * (-48 * C + 12 * C**3 + 5 * C * np.pi**2) * eps2(N)**2 * eps3(N) + (1 / 12) * (-8 - 72 * C - 12 * C**2 + 8 * C**3 + np.pi**2 + 6 * C * np.pi**2 - 8 * zeta(3)) * eps1(N) * eps2(N) * eps3(N))


def as13(N, e4= True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4 = 0
    return (-2 * eps1(N) + 2 * (2 * C + 1) * eps1(N)**2 - eps2(N) + (2 * C - 1) * eps1(N) * eps2(N) + C * eps2(N)**2 - C * eps2(N) * eps3(N)- (1 / 8) * (-8 + 4 * C**2 + np.pi**2) * (8 * eps1(N)**3 + eps2(N)**3) - (2 / 3) * (-9 - 9 * C + np.pi**2) * eps1(N)**2 * eps2(N) - (1 / 4) * (-4 - 4 * C + 4 * C**2 + np.pi**2) * eps1(N) * eps2(N)**2 + (1 / 2) * (-12 - 4 * C + 4 * C**2 + np.pi**2) * eps1(N) * eps2(N) * eps3(N) + (1 / 24) * (-12 * C**2 + np.pi**2) * (eps2(N) * eps3(N)**2 + eps2(N) * eps3(N) * epsi4) + (1 / 24) * (-48 + 36 * C**2 + 5 * np.pi**2) * eps2(N)**2 * eps3(N))


def as23(N, e4= True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4 = 0
    return  (4 * eps1(N)**2 + 2 * eps1(N) * eps2(N) + 6 * eps1(N)**2 * eps2(N) + eps2(N)**2 - eps2(N) * eps3(N)  + (1 - 2 * C) * (eps1(N) * eps2(N)**2 - 2 * eps1(N) * eps2(N) * eps3(N))- C * (8 * eps1(N)**3 + eps2(N)**3 - 3 * eps2(N)**2 * eps3(N) + eps2(N) * eps3(N)**2 + eps2(N) * eps3(N) * epsi4))


def as33(N, e4= True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4 = 0
    return (-8 * eps1(N)**3 - 2 * eps1(N) * eps2(N)**2 - eps2(N)**3 + 4 * eps1(N) * eps2(N) * eps3(N)  + 3 * eps2(N)**2 * eps3(N) - eps2(N) * eps3(N)**2 - eps2(N) * eps3(N) * epsi4)


def at03(N):
    return (1 + (-2 - 2 * C) * eps1(N) + (1 / 2) * (-6 + 4 * C + 4 * C**2 + np.pi**2) * eps1(N)**2   + (1 / 12) * (-24 - 24 * C - 12 * C**2 + np.pi**2) * eps1(N) * eps2(N) - (1 / 3) * (-16 - 24 * C + 4 * C**3 + 3 * C * np.pi**2 + 14 * zeta(3)) * eps1(N)**3  + (1 / 12) * (-96 - 72 * C + 36 * C**2 + 24 * C**3 + 13 * np.pi**2 + 10 * C * np.pi**2) * eps1(N)**2 * eps2(N)   - (1 / 12) * (8 + 24 * C + 12 * C**2 + 4 * C**3 - np.pi**2 - C * np.pi**2 + 8 * zeta(3)) * (eps1(N) * eps2(N)**2 + eps1(N) * eps2(N) * eps3(N)))


def at13(N):
    return  (-2 * eps1(N) + 2 * eps1(N)**2 + 4 * C * eps1(N)**2 + (-2 - 2 * C) * eps1(N) * eps2(N) - (-8 + 4 * C**2 + np.pi**2) * eps1(N)**3 + (1 / 6) * (-36 + 36 * C + 36 * C**2 + 5 * np.pi**2) * eps1(N)**2 * eps2(N) + (1 / 12) * (-24 - 24 * C - 12 * C**2 + np.pi**2) * (eps1(N) * eps2(N)**2 + eps1(N) * eps2(N) * eps3(N)))


def at23(N):
    return (4 * eps1(N)**2 - 8 * C * eps1(N)**3 - 2 * eps1(N) * eps2(N) + 2 * (3 + 6 * C) * eps1(N)**2 * eps2(N) - 2 * (C + 1) * (eps1(N) * eps2(N)**2 + 2 * eps1(N) * eps2(N) * eps3(N)))


def at33(N):
    return (-8 * eps1(N)**3 + 12 * eps1(N)**2 * eps2(N) - 2 * eps1(N) * eps2(N)**2 - 2 * eps1(N) * eps2(N) * eps3(N))
```

```python
#Third order scalar PPS

def PS_s3(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    PS_s_val3  = P0_s3*(as03(deltaN_star) + as13(deltaN_star)*log + 0.5*as23(deltaN_star)*log_sq + (1/6)*as33(deltaN_star)*log_cub)
    return PS_s_val3

#Third order scalar PPS with eps4=0

def PS_s3_noeps4(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    PS_s_val3  = P0_s3_noep4*(as03(deltaN_star, False) + as13(deltaN_star,False)*log + 0.5*as23(deltaN_star,False)*log_sq + (1/6)*as33(deltaN_star,False)*log_cub)
    return PS_s_val3
```

## Analytical PPS with $b$ coefficients

Now we want to compute the scalar and tensor PPS again, but using their logarithmic expansion, i.e. using the $b_S$ and $b_T$ coefficients. 

```python
#Defining the b parameters at the first order 

def bs01(N):
    return -2*(C+1)*eps1(N) - C*eps2(N)

def bs11(N):
    return -2*eps1(N) - eps2(N) 

def bs21(N):
    return 0

def bt01(N): 
    return -2*(C+1)*eps1(N)

def bt11(N):
    return -2*eps1(N)

def bt21(N):
    return 0
```

```python
#I order scalar PPS with b

def loggPS_s1(ka):
    log = np.log(ka/k_star)
    loggPS_s_val1 = P0_s1*np.exp((bs01(deltaN_star) + bs11(deltaN_star)*log))
    return loggPS_s_val1
```

```python
#Defining the b parameters at second order 

def bs02(N): 
    return -2*(C+1)*eps1(N) - C*eps2(N) + (-2*C + 0.5*np.pi**2 -5)*eps1(N)**2 + (-C**2 - 3*C + 7/12* np.pi**2 -6)*eps1(N)*eps2(N) + ( np.pi**2/8 -1)*eps2(N)**2 + (-0.5*C**2 + np.pi**2/24)*eps2(N)*eps3(N)

def bs12(N): 
    return -2*eps1(N) - eps2(N) + -2*eps1(N)**2 - (2*C+3)*eps1(N)*eps2(N)  - C*eps2(N)*eps3(N)

def bs22(N):
    return -2*eps1(N)*eps2(N)-eps2(N)*eps3(N)

def bt02(N):
    return -2*(C+1)*eps1(N) + (-2*C + 0.5*np.pi**2 -5)*eps1(N)**2 + (-C**2 -2*C + np.pi**2/12 -2)*eps1(N)*eps2(N)

def bt12(N):
    return -2*eps1(N) -2*eps1(N)**2- 2*(C+1)*eps1(N)*eps2(N)

def bt22(N):
    return - 2*eps1(N)*eps2(N)
```

```python
#II order scalar PPS with b

def loggPS_s2(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    loggPS_s_val2  = P0_s2*np.exp((bs02(deltaN_star) + bs12(deltaN_star)*log + 0.5*bs22(deltaN_star)*log_sq))
    return loggPS_s_val2
```

```python
#III order scalar PPS with b

def loggPS_s3(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    loggPS_s_val3  = P0_s3*np.exp((bs03(deltaN_star) + bs13(deltaN_star)*log + 0.5*bs23(deltaN_star)*log_sq+(1/6)*bs33(deltaN_star)*log_cub))
    return loggPS_s_val3

#III order scalar PPS with b with epsilon_4=0

def loggPS_s3_noeps4(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    loggPS_s_val3  = P0_s3_noep4*np.exp((bs03(deltaN_star, False) + bs13(deltaN_star, False)*log + 0.5*bs23(deltaN_star, False)*log_sq+(1/6)*bs33(deltaN_star, False)*log_cub))
    return loggPS_s_val3
```

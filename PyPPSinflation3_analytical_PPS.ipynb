By approximating at leading order (LO) $H^2$ as

$H^2 \simeq \frac{1}{3M^2_{Pl}} V,$

it is possible to find

$\frac{d}{dN}\Bigg|^{LO} = -M^2_{\text{Pl}} \frac{V_\phi}{V}\frac{d}{d\phi},$


and from it the $\varepsilon_i$ at leading order are found to be analytically:


$\varepsilon_1^{LO} \simeq \frac{M^2_{Pl}}{2} \left(\frac{V'}{V}\right)^2,$

$\varepsilon_2^{LO} \simeq 2M^2_{Pl}\left[\left(\frac{V'}{V}\right)^2-\frac{V''}{V}\right],$

$\varepsilon_2^{LO} \varepsilon_3^{LO} \simeq 2M^4_{Pl}\left[\frac{V''' V'}{V^2}-\frac{3V'' V'^2}{V^3} +2\left(\frac{V'}{V}\right)^4\right],$

$\epsilon_4^{LO} = \frac{2 M_{\text{Pl}}^6}{\epsilon_2 \epsilon_3} \bigg[\frac{8 V_\phi^6}{V^6} - \frac{17 V_\phi^4 V_{\phi\phi}}{V^5} + \frac{6 V_\phi^2 V_{\phi\phi}^2}{V^4} + \frac{5 V_\phi^3 V_{\phi\phi\phi}}{V^4} - \frac{V_\phi V_{\phi\phi} V_{\phi\phi\phi}}{V^3} - \frac{V_\phi^2 V_{\phi\phi\phi\phi}}{V^3}  + \frac{2 M_{\text{Pl}}^2}{\epsilon_2} \bigg( - \frac{4 V_\phi^8}{V^8} + \frac{12 V_\phi^6 V_{\phi\phi}}{V^7} - \frac{9 V_\phi^4 V_{\phi\phi}^2}{V^6} - \frac{4 V_\phi^5 V_{\phi\phi\phi}}{V^6} + \frac{6 V_\phi^3 V_{\phi\phi} V_{\phi\phi\phi}}{V^5}  - \frac{V_\phi^2 V_{\phi\phi\phi}^2}{V^4} \bigg) \bigg].$


These equations can be developed at higher order by searching for the equations at next to the leading order (NLO). Starting from the Friedmann equation and the **exact** equations

$H^2 = \frac{V}{3M_{\text{Pl}}} \left(1-\frac{\varepsilon_1}{3}\right)^{-1} \ \ \ \text{and} \ \ \ dN = \pm \frac{1}{M_{\text{Pl}}} \frac{d\phi}{\sqrt{2\varepsilon_1}},$

it is possibile to find 

$\varepsilon_0 = \varepsilon_0^{LO} \sqrt{1-\frac{\varepsilon_1}{3}}$

and

$\frac{d}{dN}\Bigg|^{NLO} = \sqrt{\frac{\varepsilon_1^{NLO}}{\varepsilon_1^{LO}}} \frac{d}{dN}\Bigg|^{LO}.$

Using the latter equations it is possible to find the the $\varepsilon_i$ at higher orders. If we expand up to a certain order, it is possible to reach up to the next-next-next-leading-order (N3LO):

$\varepsilon_1^{N3LO} = \varepsilon_1^{LO} - \frac{1}{3} \varepsilon_1^{LO} \varepsilon_2^{LO} - \frac{1}{9} \varepsilon_1^{LO^{2}} \varepsilon_2^{LO} + \frac{5}{36} \varepsilon_1^{LO} \varepsilon_2^{LO^{2}} + \frac{1}{9} \varepsilon_1^{LO} \varepsilon_2^{LO} \varepsilon_3^{LO}$

For more details, see V. Vennin - Horizon-Flow off-track for Inflation (https://arxiv.org/abs/1401.2926) and J. Martin et. al. - Cosmic Inflation at the Crossroads (https://arxiv.org/abs/2404.10647).

The analytical PPS are compute both at LO and N3LO, where we take the N3LO spectra as PPS to be compared with the numerical results. 

 
-----

For the background and its relation with the e-folds number, we have: 

$N_{end}- N \simeq -\frac{1}{M^2_{Pl}} \int_{\phi}^{\phi_{end}} \frac{V}{V'} d\phi$,

where we define $\phi_{end}$ to be the value when $\varepsilon_1=1$. 

-----

We want to follow these steps:

1. Define the potential, i.e., define the model.

2. Compute $V$, $V'$, $V''$, $\varepsilon_1$, $\varepsilon_2$, $\varepsilon_3$, and $\varepsilon_4$.

3. Compute $\phi_{end}$ from $\varepsilon_1(\phi_{end})=1$.

4. Compute $\Delta N$.

5. Invert the result from the previous step to find $\phi(\Delta N)$.

6. Rewrite the expressions in terms of $\Delta N$.

7. Compute $a_{S0,1,2}$ and $a_{T0,1,2}$.

8. Compute $\mathcal{P}(k)$.

Note that at first order, terms like $\varepsilon^2_1, \varepsilon^2_2, \varepsilon_2\varepsilon_3$, etc., are set to zero, simplifying the equations for $a_{S/T}$. At second order, the equations are complete. In Vennin's paper, the approach differs as he modifies the equations for $H$ and $\varepsilon_i$.

Steps 1 to 6 should be computed analytically by hand, or at most by using analytical software such as Mathematica or Maple. The PyPPSinflation package includes a Wolfram Mathematica notebook, Quantities_wrt_N, which can compute steps 1 to 6 analytically for any inflation model. Steps 7 and 8 are computed in this Jupyter notebook.

## An example of analytical computation with the simplest model: $\frac{1}{2}m^2 \phi^2$.

In this notebook, we use the KKLT model as an example. Since its equations are too lengthy, we also present the simplest model of inflation for computation examples.

For the $V(\phi) = \frac{1}{2}m^2\phi^2$ model, performing all the calculations (which are simple enough to be done by hand) yields:


$H^2 = \frac{m^2 \phi^2}{6 M^2_{Pl}}$

$\varepsilon_1 =\frac{ 2M^2_{Pl}}{\phi^2}$

$\varepsilon_2 =\varepsilon_3= 2\varepsilon_1$

From the equation $\varepsilon_1(\phi_{end}) =1$ we get $\phi^2_{end} = 2 M^2_{Pl}$, and integrating the $\Delta N$ and inverting it we get

$\phi^2 = 4 M^2_{Pl} \left(\Delta N + \frac{1}{2} \right).$

Substituting we have

$\varepsilon_1  = \frac{\varepsilon_2}{2}= \frac{\varepsilon_3}{2}=\frac{1}{2\Delta N +1} \ \ \ and \ \ \ H^2 = \frac{2}{3} m^2(\Delta N+\frac{1}{2})$,

and then we are able to easely compute the PPS. 

## The choice of the coupling constant

In the first notebook of PyPPSinflation, which is dedicated to numerical computation, we explain the procedure for normalization, specifically the choice of the coupling constant. Here, we focus on the second step of that procedure, which concerns normalization for the analytical part.

Let's recall some important points:

We choose the coupling constant of the theory, which in this case is the mass, by imposing that the scalar primordial power spectrum at the pivot scale has the value $\mathscr{P}_{R}(k_*) = a_0 \mathscr{P}_{0R}(k_*) = 2\times 10^{-9}$.

To achieve the same value of the PPS at the pivot scale for both numerical and analytical computations, it is possible to use different coupling constants.

At the end of this notebook, we will address the normalization of the analytical coupling constant(s).

```python
#Defining the Hubble function at the second order
#(it is always the same at every order, we do it for the sake of completness)

def H_sq2(mass,N):
    H_sq2 = 1/3 * (mass)**2 * ( ( 1 + 2 * mu * ( ( ( ( 12 * N * mu + ( ( ( 2 )**( \
1/2 ) + -1 * mu ) * mu * ( mu * ( 2 * ( 2 )**( 1/2 ) + mu ) )**( 1/2 \
) + ( 2 )**( 1/2 ) * ( ( mu )**( 2 ) * ( ( 2 * ( 2 )**( 1/2 ) + -3 * \
mu ) * mu + 12 * N * ( 6 * N + ( ( 2 )**( 1/2 ) + -1 * mu ) \
* ( mu * ( 2 * ( 2 )**( 1/2 ) + mu ) )**( 1/2 ) ) ) )**( 1/2 ) ) ) \
)**( 1/3 ) + mu * ( -1 + mu * ( ( 12 * N * mu + ( ( ( 2 )**( 1/2 \
) + -1 * mu ) * mu * ( mu * ( 2 * ( 2 )**( 1/2 ) + mu ) )**( 1/2 ) + \
( 2 )**( 1/2 ) * ( ( mu )**( 2 ) * ( ( 2 * ( 2 )**( 1/2 ) + -3 * mu ) \
* mu + 12 * N * ( 6 * N + ( ( 2 )**( 1/2 ) + -1 * mu ) * ( \
mu * ( 2 * ( 2 )**( 1/2 ) + mu ) )**( 1/2 ) ) ) )**( 1/2 ) ) ) )**( \
-1/3 ) ) ) )**( -1 ) ) )**( -1 )
    return H_sq2
```

```python
#The P0s at the second order
#(they are always the same at every order, we do it for the sake of completness)

P0_s2 = H_sq2(m2an_s,deltaN_star)/(8*np.pi**2 * Mpl**2 *  eps1(deltaN_star))

P0_t2 = 2*H_sq2(m2an_t, deltaN_star)/(np.pi**2 * Mpl**2)
```

```python
#Defining the a parameters at the second order 

C = -0.7296

def as02(N): 
    return 1-2*(C+1)*eps1(N) - C*eps2(N) + (2*C**2 +2*C + 0.5*np.pi**2 -3)*eps1(N)**2 + (C**2 - C + 7/12* np.pi**2 -6)*eps1(N)*eps2(N) + (0.5*C**2 + np.pi**2/8 -1)*eps2(N)**2 + (-0.5*C**2 + np.pi**2/24)*eps2(N)*eps3(N)

def as12(N): 
    return -2*eps1(N) - eps2(N) + 2*(2*C+1)*eps1(N)**2 + (2*C-1)*eps1(N)*eps2(N) + C*eps2(N)**2 - C*eps2(N)*eps3(N)

def as22(N):
    return 4*eps1(N)**2 +2*eps1(N)*eps2(N)+eps2(N)**2-eps2(N)*eps3(N)

def at02(N):
    return 1-2*(C+1)*eps1(N) + (2*C**2 +2*C + 0.5*np.pi**2 -3)*eps1(N)**2 + (-C**2 -2*C + np.pi**2/12 -2)*eps1(N)*eps2(N)

def at12(N):
    return -2*eps1(N) +2*(2*C+1)*eps1(N)**2- 2*(C+1)*eps1(N)*eps2(N)

def at22(N):
    return  4*eps1(N)**2 - 2*eps1(N)*eps2(N)
```

```python
#Scalar PPS at the second order

def PS_s2(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    PS_s_val2  = P0_s2*(as02(deltaN_star) + as12(deltaN_star)*log + 0.5*as22(deltaN_star)*log_sq)
    return PS_s_val2
```

```python
#Defining the Hubble function at the third order
#(it is always the same at every order, we do it for the sake of completness)

def H_sq3(mass,N):
    H_sq3 = 1/3 * (mass)**2 * ( ( 1 + 2 * mu * ( ( ( ( 12 * N * mu + ( ( ( 2 )**( \
1/2 ) + -1 * mu ) * mu * ( mu * ( 2 * ( 2 )**( 1/2 ) + mu ) )**( 1/2 \
) + ( 2 )**( 1/2 ) * ( ( mu )**( 2 ) * ( ( 2 * ( 2 )**( 1/2 ) + -3 * \
mu ) * mu + 12 * N * ( 6 * N + ( ( 2 )**( 1/2 ) + -1 * mu ) \
* ( mu * ( 2 * ( 2 )**( 1/2 ) + mu ) )**( 1/2 ) ) ) )**( 1/2 ) ) ) \
)**( 1/3 ) + mu * ( -1 + mu * ( ( 12 * N * mu + ( ( ( 2 )**( 1/2 \
) + -1 * mu ) * mu * ( mu * ( 2 * ( 2 )**( 1/2 ) + mu ) )**( 1/2 ) + \
( 2 )**( 1/2 ) * ( ( mu )**( 2 ) * ( ( 2 * ( 2 )**( 1/2 ) + -3 * mu ) \
* mu + 12 * N * ( 6 * N + ( ( 2 )**( 1/2 ) + -1 * mu ) * ( \
mu * ( 2 * ( 2 )**( 1/2 ) + mu ) )**( 1/2 ) ) ) )**( 1/2 ) ) ) )**( \
-1/3 ) ) ) )**( -1 ) ) )**( -1 )
    return H_sq3
```

```python
#The P0s at the third order
#(they are always the same at every order, we do it for the sake of completness)

P0_s3 = H_sq3(m2an_s,deltaN_star)/(8*np.pi**2 * Mpl**2 *  eps1(deltaN_star))

P0_s3_noep4 = H_sq3(m2an_s_noeps4,deltaN_star)/(8*np.pi**2 * Mpl**2 *  eps1(deltaN_star)) #

P0_t3 = 2*H_sq3(m2an_t, deltaN_star)/(np.pi**2 * Mpl**2)
```

```python
#Defining the a parameters at the third order 
#The scalar coefficients are defined so that it is possible to set eps4(N) = 0

C = -0.7296

def as03(N, e4 = True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4 = 0
    return (1 - 2 * (1 + C) * eps1(N) + (-3 + 2 * C + 2 * C**2 + (np.pi**2) / 2) * eps1(N)**2 - C * eps2(N) + (-6 - C + C**2 + (7 * np.pi**2) / 12) * eps1(N) * eps2(N) + (1 / 8) * (-8 + 4 * C**2 + np.pi**2) * eps2(N)**2 + (1 / 24) * (-12 * C**2 + np.pi**2) * eps2(N) * eps3(N) - (1 / 24) * (-16 - 24 * C + 4 * C**3 + 3 * C * np.pi**2 + 14 * zeta(3)) * (8 * eps1(N)**3 + eps2(N)**3) + (1 / 12) * (72 * C + 36 * C**2 + 13 * np.pi**2 - 8 * C * np.pi**2 - 84 * zeta(3)) * eps1(N)**2 * eps2(N) - (1 / 24) * (16 - 24 * C - 12 * C**2 + 8 * C**3 - 15 * np.pi**2 + 6 * C * np.pi**2 + 100 * zeta(3)) * eps1(N) * eps2(N)**2 + (1 / 24) * (16 - 4 * C**3 + C * np.pi**2 - 8 * zeta(3)) * (eps2(N) * eps3(N)**2 + eps2(N) * eps3(N) * epsi4) + (1 / 24) * (-48 * C + 12 * C**3 + 5 * C * np.pi**2) * eps2(N)**2 * eps3(N) + (1 / 12) * (-8 - 72 * C - 12 * C**2 + 8 * C**3 + np.pi**2 + 6 * C * np.pi**2 - 8 * zeta(3)) * eps1(N) * eps2(N) * eps3(N))


def as13(N, e4= True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4 = 0
    return (-2 * eps1(N) + 2 * (2 * C + 1) * eps1(N)**2 - eps2(N) + (2 * C - 1) * eps1(N) * eps2(N) + C * eps2(N)**2 - C * eps2(N) * eps3(N)- (1 / 8) * (-8 + 4 * C**2 + np.pi**2) * (8 * eps1(N)**3 + eps2(N)**3) - (2 / 3) * (-9 - 9 * C + np.pi**2) * eps1(N)**2 * eps2(N) - (1 / 4) * (-4 - 4 * C + 4 * C**2 + np.pi**2) * eps1(N) * eps2(N)**2 + (1 / 2) * (-12 - 4 * C + 4 * C**2 + np.pi**2) * eps1(N) * eps2(N) * eps3(N) + (1 / 24) * (-12 * C**2 + np.pi**2) * (eps2(N) * eps3(N)**2 + eps2(N) * eps3(N) * epsi4) + (1 / 24) * (-48 + 36 * C**2 + 5 * np.pi**2) * eps2(N)**2 * eps3(N))


def as23(N, e4= True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4 = 0
    return  (4 * eps1(N)**2 + 2 * eps1(N) * eps2(N) + 6 * eps1(N)**2 * eps2(N) + eps2(N)**2 - eps2(N) * eps3(N)  + (1 - 2 * C) * (eps1(N) * eps2(N)**2 - 2 * eps1(N) * eps2(N) * eps3(N))- C * (8 * eps1(N)**3 + eps2(N)**3 - 3 * eps2(N)**2 * eps3(N) + eps2(N) * eps3(N)**2 + eps2(N) * eps3(N) * epsi4))


def as33(N, e4= True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4 = 0
    return (-8 * eps1(N)**3 - 2 * eps1(N) * eps2(N)**2 - eps2(N)**3 + 4 * eps1(N) * eps2(N) * eps3(N)  + 3 * eps2(N)**2 * eps3(N) - eps2(N) * eps3(N)**2 - eps2(N) * eps3(N) * epsi4)


def at03(N):
    return (1 + (-2 - 2 * C) * eps1(N) + (1 / 2) * (-6 + 4 * C + 4 * C**2 + np.pi**2) * eps1(N)**2   + (1 / 12) * (-24 - 24 * C - 12 * C**2 + np.pi**2) * eps1(N) * eps2(N) - (1 / 3) * (-16 - 24 * C + 4 * C**3 + 3 * C * np.pi**2 + 14 * zeta(3)) * eps1(N)**3  + (1 / 12) * (-96 - 72 * C + 36 * C**2 + 24 * C**3 + 13 * np.pi**2 + 10 * C * np.pi**2) * eps1(N)**2 * eps2(N)   - (1 / 12) * (8 + 24 * C + 12 * C**2 + 4 * C**3 - np.pi**2 - C * np.pi**2 + 8 * zeta(3)) * (eps1(N) * eps2(N)**2 + eps1(N) * eps2(N) * eps3(N)))


def at13(N):
    return  (-2 * eps1(N) + 2 * eps1(N)**2 + 4 * C * eps1(N)**2 + (-2 - 2 * C) * eps1(N) * eps2(N) - (-8 + 4 * C**2 + np.pi**2) * eps1(N)**3 + (1 / 6) * (-36 + 36 * C + 36 * C**2 + 5 * np.pi**2) * eps1(N)**2 * eps2(N) + (1 / 12) * (-24 - 24 * C - 12 * C**2 + np.pi**2) * (eps1(N) * eps2(N)**2 + eps1(N) * eps2(N) * eps3(N)))


def at23(N):
    return (4 * eps1(N)**2 - 8 * C * eps1(N)**3 - 2 * eps1(N) * eps2(N) + 2 * (3 + 6 * C) * eps1(N)**2 * eps2(N) - 2 * (C + 1) * (eps1(N) * eps2(N)**2 + 2 * eps1(N) * eps2(N) * eps3(N)))


def at33(N):
    return (-8 * eps1(N)**3 + 12 * eps1(N)**2 * eps2(N) - 2 * eps1(N) * eps2(N)**2 - 2 * eps1(N) * eps2(N) * eps3(N))
```

```python
#Third order scalar PPS

def PS_s3(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    PS_s_val3  = P0_s3*(as03(deltaN_star) + as13(deltaN_star)*log + 0.5*as23(deltaN_star)*log_sq + (1/6)*as33(deltaN_star)*log_cub)
    return PS_s_val3

#Third order scalar PPS with eps4=0

def PS_s3_noeps4(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    PS_s_val3  = P0_s3_noep4*(as03(deltaN_star, False) + as13(deltaN_star,False)*log + 0.5*as23(deltaN_star,False)*log_sq + (1/6)*as33(deltaN_star,False)*log_cub)
    return PS_s_val3
```

## Analytical PPS with $b$ coefficients

Now we want to compute the scalar and tensor PPS again, but using their logarithmic expansion, i.e. using the $b_S$ and $b_T$ coefficients. 

```python
#Defining the b parameters at the first order 

def bs01(N):
    return -2*(C+1)*eps1(N) - C*eps2(N)

def bs11(N):
    return -2*eps1(N) - eps2(N) 

def bs21(N):
    return 0

def bt01(N): 
    return -2*(C+1)*eps1(N)

def bt11(N):
    return -2*eps1(N)

def bt21(N):
    return 0
```

```python
#I order scalar PPS with b

def loggPS_s1(ka):
    log = np.log(ka/k_star)
    loggPS_s_val1 = P0_s1*np.exp((bs01(deltaN_star) + bs11(deltaN_star)*log))
    return loggPS_s_val1
```

```python
#Defining the b parameters at second order 

def bs02(N): 
    return -2*(C+1)*eps1(N) - C*eps2(N) + (-2*C + 0.5*np.pi**2 -5)*eps1(N)**2 + (-C**2 - 3*C + 7/12* np.pi**2 -6)*eps1(N)*eps2(N) + ( np.pi**2/8 -1)*eps2(N)**2 + (-0.5*C**2 + np.pi**2/24)*eps2(N)*eps3(N)

def bs12(N): 
    return -2*eps1(N) - eps2(N) + -2*eps1(N)**2 - (2*C+3)*eps1(N)*eps2(N)  - C*eps2(N)*eps3(N)

def bs22(N):
    return -2*eps1(N)*eps2(N)-eps2(N)*eps3(N)

def bt02(N):
    return -2*(C+1)*eps1(N) + (-2*C + 0.5*np.pi**2 -5)*eps1(N)**2 + (-C**2 -2*C + np.pi**2/12 -2)*eps1(N)*eps2(N)

def bt12(N):
    return -2*eps1(N) -2*eps1(N)**2- 2*(C+1)*eps1(N)*eps2(N)

def bt22(N):
    return - 2*eps1(N)*eps2(N)
```

```python
#Defining the b parameters at the third order 
#The scalar coefficients are defined so that it is possible to set eps4(N) = 0


def bs03(N, e4=True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4=0
    return (-2 * (1 + C) * eps1(N) + (-5 - 2 * C + (np.pi**2) / 2) * eps1(N)**2 - C * eps2(N) + (-6 - 3 * C - C**2 + (7 * np.pi**2) / 12) * eps1(N) * eps2(N)+ (1 / 24) * (-12 * C**2 + np.pi**2) * eps2(N) * eps3(N) - (1 / 24) * (80 + 48 * C - 24 * np.pi**2 + 112 * zeta(3)) * eps1(N)**3 + (1 / 8) * (-8 + np.pi**2) * eps2(N)**2  - (1 / 24) * (-16 + 14 * zeta(3)) * eps2(N)**3 + (1 / 12) * (-144 - 180 * C - 36 * C**2 + 27 * np.pi**2 + 12 * C * np.pi**2 - 84 * zeta(3)) * eps1(N)**2 * eps2(N)- (1 / 24) * (64 + 168 * C + 36 * C**2 + 8 * C**3 - 21 * np.pi**2 - 14 * C * np.pi**2 + 100 * zeta(3)) * eps1(N) * eps2(N)**2+ (1 / 24) * (16 - 4 * C**3 + C * np.pi**2 - 8 * zeta(3)) * (eps2(N) * eps3(N)**2 + eps2(N) * eps3(N) * epsi4)  + (1 / 24) * (-48 * C + 6 * C * np.pi**2) * eps2(N)**2 * eps3(N)+ (1 / 12) * (-8 - 72 * C - 24 * C**2 - 4 * C**3 + 2 * np.pi**2 + 7 * C * np.pi**2 - 8 * zeta(3)) * eps1(N) * eps2(N) * eps3(N))


def bs13(N, e4=True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4=0
    return (-2 * eps1(N) - 2 * eps1(N)**2 - 2 * eps1(N)**3 - eps2(N) - (3 + 2 * C) * eps1(N) * eps2(N) - C * eps2(N) * eps3(N) - (2 / 3) * (45 / 2 + 9 * C - 3 * np.pi**2 / 2) * eps1(N)**2 * eps2(N)- (1 / 4) * (28 + 12 * C + 4 * C**2 - 7 * np.pi**2 / 3) * eps1(N) * eps2(N)**2 + (1 / 2) * (-12 - 8 * C - 2 * C**2 + 7 * np.pi**2 / 6) * eps1(N) * eps2(N) * eps3(N)+ (1 / 24) * (-12 * C**2 + np.pi**2) * (eps2(N) * eps3(N)**2 + eps2(N) * eps3(N) * epsi4) + (1 / 24) * (-48 + 6 * np.pi**2) * eps2(N)**2 * eps3(N))


def bs23(N, e4=True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4=0
    return (-2 * eps1(N) * eps2(N) - 6 * eps1(N)**2 * eps2(N) - (3 + 2 * C) * eps1(N) * eps2(N)**2 - eps2(N) * eps3(N)- 2 * (2 + C) * eps1(N) * eps2(N) * eps3(N) - C * eps2(N) * eps3(N)**2 - C * eps2(N) * eps3(N) * epsi4)


def bs33(N, e4=True):
    if e4:
        epsi4 = eps4(N)
    else:
        epsi4=0
    return  (-2 * eps1(N) * eps2(N)**2 - 2 * eps1(N) * eps2(N) * eps3(N) - eps2(N) * eps3(N)**2 - eps2(N) * eps3(N) * epsi4)


def bt03(N):
    return (-2 * (1 + C) * eps1(N) + (1 / 2) * (-10 - 4 * C + np.pi**2) * eps1(N)**2  + (1 / 12) * (-24 - 24 * C - 12 * C**2 + np.pi**2) * eps1(N) * eps2(N) - (1 / 3) * (10 + 6 * C - 3 * np.pi**2 + 14 * zeta(3)) * eps1(N)**3  + (1 / 12) * (-144 - 168 * C - 36 * C**2 + 15 * np.pi**2 + 12 * C * np.pi**2) * eps1(N)**2 * eps2(N) - (1 / 12) * (8 + 24 * C + 12 * C**2 + 4 * C**3 - np.pi**2 - C * np.pi**2 + 8 * zeta(3)) * (eps1(N) * eps2(N)**2 + eps1(N) * eps2(N) * eps3(N)))


def bt13(N):
    return (-2 * eps1(N) - 2 * eps1(N)**2 + (-2 - 2 * C) * eps1(N) * eps2(N) - 2 * eps1(N)**3  + (1 / 6) * (-84 - 36 * C + 6 * np.pi**2) * eps1(N)**2 * eps2(N) + (1 / 12) * (-24 - 24 * C - 12 * C**2 + np.pi**2) * (eps1(N) * eps2(N)**2 + eps1(N) * eps2(N) * eps3(N)))


def bt23(N):
    return (-2 * eps1(N) * eps2(N) - 6 * eps1(N)**2 * eps2(N) - 2 * (1 + C) * eps1(N) * eps2(N)**2 - 2 * (1 + C) * eps1(N) * eps2(N) * eps3(N))

def bt33(N):
    return (-2 * eps1(N) * eps2(N)**2 - 2 * eps1(N) * eps2(N) * eps3(N))
```

```python
#III order scalar PPS with b

def loggPS_s3(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    loggPS_s_val3  = P0_s3*np.exp((bs03(deltaN_star) + bs13(deltaN_star)*log + 0.5*bs23(deltaN_star)*log_sq+(1/6)*bs33(deltaN_star)*log_cub))
    return loggPS_s_val3

#III order scalar PPS with b with epsilon_4=0

def loggPS_s3_noeps4(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    loggPS_s_val3  = P0_s3_noep4*np.exp((bs03(deltaN_star, False) + bs13(deltaN_star, False)*log + 0.5*bs23(deltaN_star, False)*log_sq+(1/6)*bs33(deltaN_star, False)*log_cub))
    return loggPS_s_val3
```

```python
#Defining the b parameters at the third order 
#The scalar coefficients are defined so that it is possible to set eps4(N) = 0


def bs03_LO(N, e4_LO=True):
    if e4_LO:
        epsi4_LO = eps4_LO(N)
    else:
        epsi4_LO=0
    return (-2 * (1 + C) * eps1_LO(N) + (-5 - 2 * C + (np.pi**2) / 2) * eps1_LO(N)**2 - C * eps2_LO(N) + (-6 - 3 * C - C**2 + (7 * np.pi**2) / 12) * eps1_LO(N) * eps2_LO(N)+ (1 / 24) * (-12 * C**2 + np.pi**2) * eps2_LO(N) * eps3_LO(N) - (1 / 24) * (80 + 48 * C - 24 * np.pi**2 + 112 * zeta(3)) * eps1_LO(N)**3 + (1 / 8) * (-8 + np.pi**2) * eps2_LO(N)**2  - (1 / 24) * (-16 + 14 * zeta(3)) * eps2_LO(N)**3 + (1 / 12) * (-144 - 180 * C - 36 * C**2 + 27 * np.pi**2 + 12 * C * np.pi**2 - 84 * zeta(3)) * eps1_LO(N)**2 * eps2_LO(N)- (1 / 24) * (64 + 168 * C + 36 * C**2 + 8 * C**3 - 21 * np.pi**2 - 14 * C * np.pi**2 + 100 * zeta(3)) * eps1_LO(N) * eps2_LO(N)**2+ (1 / 24) * (16 - 4 * C**3 + C * np.pi**2 - 8 * zeta(3)) * (eps2_LO(N) * eps3_LO(N)**2 + eps2_LO(N) * eps3_LO(N) * epsi4_LO)  + (1 / 24) * (-48 * C + 6 * C * np.pi**2) * eps2_LO(N)**2 * eps3_LO(N)+ (1 / 12) * (-8 - 72 * C - 24 * C**2 - 4 * C**3 + 2 * np.pi**2 + 7 * C * np.pi**2 - 8 * zeta(3)) * eps1_LO(N) * eps2_LO(N) * eps3_LO(N))


def bs13_LO(N, e4_LO=True):
    if e4_LO:
        epsi4_LO = eps4_LO(N)
    else:
        epsi4_LO=0
    return (-2 * eps1_LO(N) - 2 * eps1_LO(N)**2 - 2 * eps1_LO(N)**3 - eps2_LO(N) - (3 + 2 * C) * eps1_LO(N) * eps2_LO(N) - C * eps2_LO(N) * eps3_LO(N) - (2 / 3) * (45 / 2 + 9 * C - 3 * np.pi**2 / 2) * eps1_LO(N)**2 * eps2_LO(N)- (1 / 4) * (28 + 12 * C + 4 * C**2 - 7 * np.pi**2 / 3) * eps1_LO(N) * eps2_LO(N)**2 + (1 / 2) * (-12 - 8 * C - 2 * C**2 + 7 * np.pi**2 / 6) * eps1_LO(N) * eps2_LO(N) * eps3_LO(N)+ (1 / 24) * (-12 * C**2 + np.pi**2) * (eps2_LO(N) * eps3_LO(N)**2 + eps2_LO(N) * eps3_LO(N) * epsi4_LO) + (1 / 24) * (-48 + 6 * np.pi**2) * eps2_LO(N)**2 * eps3_LO(N))


def bs23_LO(N, e4_LO=True):
    if e4_LO:
        epsi4_LO = eps4_LO(N)
    else:
        epsi4_LO=0
    return (-2 * eps1_LO(N) * eps2_LO(N) - 6 * eps1_LO(N)**2 * eps2_LO(N) - (3 + 2 * C) * eps1_LO(N) * eps2_LO(N)**2 - eps2_LO(N) * eps3_LO(N)- 2 * (2 + C) * eps1_LO(N) * eps2_LO(N) * eps3_LO(N) - C * eps2_LO(N) * eps3_LO(N)**2 - C * eps2_LO(N) * eps3_LO(N) * epsi4_LO)


def bs33_LO(N, e4_LO=True):
    if e4_LO:
        epsi4_LO = eps4_LO(N)
    else:
        epsi4_LO=0
    return  (-2 * eps1_LO(N) * eps2_LO(N)**2 - 2 * eps1_LO(N) * eps2_LO(N) * eps3_LO(N) - eps2_LO(N) * eps3_LO(N)**2 - eps2_LO(N) * eps3_LO(N) * epsi4_LO)


def bt03_LO(N):
    return (-2 * (1 + C) * eps1_LO(N) + (1 / 2) * (-10 - 4 * C + np.pi**2) * eps1_LO(N)**2  + (1 / 12) * (-24 - 24 * C - 12 * C**2 + np.pi**2) * eps1_LO(N) * eps2_LO(N) - (1 / 3) * (10 + 6 * C - 3 * np.pi**2 + 14 * zeta(3)) * eps1_LO(N)**3  + (1 / 12) * (-144 - 168 * C - 36 * C**2 + 15 * np.pi**2 + 12 * C * np.pi**2) * eps1_LO(N)**2 * eps2_LO(N) - (1 / 12) * (8 + 24 * C + 12 * C**2 + 4 * C**3 - np.pi**2 - C * np.pi**2 + 8 * zeta(3)) * (eps1_LO(N) * eps2_LO(N)**2 + eps1_LO(N) * eps2_LO(N) * eps3_LO(N)))


def bt13_LO(N):
    return (-2 * eps1_LO(N) - 2 * eps1_LO(N)**2 + (-2 - 2 * C) * eps1_LO(N) * eps2_LO(N) - 2 * eps1_LO(N)**3  + (1 / 6) * (-84 - 36 * C + 6 * np.pi**2) * eps1_LO(N)**2 * eps2_LO(N) + (1 / 12) * (-24 - 24 * C - 12 * C**2 + np.pi**2) * (eps1_LO(N) * eps2_LO(N)**2 + eps1_LO(N) * eps2_LO(N) * eps3_LO(N)))


def bt23_LO(N):
    return (-2 * eps1_LO(N) * eps2_LO(N) - 6 * eps1_LO(N)**2 * eps2_LO(N) - 2 * (1 + C) * eps1_LO(N) * eps2_LO(N)**2 - 2 * (1 + C) * eps1_LO(N) * eps2_LO(N) * eps3_LO(N))

def bt33_LO(N):
    return (-2 * eps1_LO(N) * eps2_LO(N)**2 - 2 * eps1_LO(N) * eps2_LO(N) * eps3_LO(N))
```

```python
#The P0s at the third order
#(they are always the same at every order, we do it for the sake of completness)

P0_s3_LO = H_sq3(m2an_s_LO,deltaN_star)/(8*np.pi**2 * Mpl**2 *  eps1_LO(deltaN_star))

P0_s3_noep4_LO = H_sq3(m2an_s_noeps4_LO,deltaN_star)/(8*np.pi**2 * Mpl**2 *  eps1_LO(deltaN_star)) #

P0_t3_LO = 2*H_sq3(m2an_t_LO, deltaN_star)/(np.pi**2 * Mpl**2)
```

```python
#III order scalar PPS with b

def loggPS_s3_LO(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    loggPS_s_val3_LO  = P0_s3_LO*np.exp((bs03_LO(deltaN_star) + bs13_LO(deltaN_star)*log + 0.5*bs23_LO(deltaN_star)*log_sq+(1/6)*bs33_LO(deltaN_star)*log_cub))
    return loggPS_s_val3_LO

#III order scalar PPS with b with epsilon_4=0

def loggPS_s3_noeps4_LO(ka): 
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    loggPS_s_val3_LO  = P0_s3_noep4_LO*np.exp((bs03_LO(deltaN_star, False) + bs13_LO(deltaN_star, False)*log + 0.5*bs23_LO(deltaN_star, False)*log_sq+(1/6)*bs33_LO(deltaN_star, False)*log_cub))
    return loggPS_s_val3_LO
```

```python
#III order tensor PPS with b

def loggPS_t3_LO(ka):
    log = np.log(ka/k_star)
    log_sq = log**2
    log_cub = log**3
    loggPS_t_val3_LO  = P0_t3_LO*np.exp((bt03_LO(deltaN_star) + bt13_LO(deltaN_star)*log + 0.5*bt23_LO(deltaN_star)*log_sq+(1/6)*bt33_LO(deltaN_star)*log_cub))
    return loggPS_t_val3_LO
```
